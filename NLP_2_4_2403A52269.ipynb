```python
import nltk
import spacy
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer

nltk.download('punkt')
```

```python
!python -m spacy download en_core_web_sm
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text="""A patient is suffering from a thyroid a common medical condition more prevalent in women refers to disorders where the buttterfly-shaped thyroid gland in the neck produces too little harmone or too much."""
```

```python
print(text)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
a=sent_tokenize(text)
b=word_tokenize(text)

print(a)
print(b)
```

```python
doc = nlp(text)

sentences_spacy = [sent.text for sent in doc.sents]
words_spacy = [token.text for token in doc]

print(sentences_spacy)
print(words_spacy)
```

```python
ps = PorterStemmer()

stemmed_words = [ps.stem(word) for word in b]

print("Original Words:\n", b[:20])
print("\nStemmed Words:\n", stemmed_words[:20])
```

```python
lemmatized_words = [token.lemma_ for token in doc]

print("Original Words:\n", words_spacy[:20])
print("\nLemmatized Words:\n", lemmatized_words[:20])
```

```python
for word in ["patients", "diagnosis", "affects", "monitoring", "complications"]:
    print(f"Word: {word}")
    print(" Stemming:", ps.stem(word))
    token = nlp(word)[0]
    print(" Lemmatization:", token.lemma_)
    print()
```
