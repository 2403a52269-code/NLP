```python
import pandas as pd

data = {
    'text': [
        "Congratulations you won a lottery",
        "Claim your free gift card now",
        "Exclusive deal just for you",
        "Good morning have a nice day",
        "Can we reschedule our meeting",
        "Did you finish the assignment",
        "Get cash bonus instantly",
        "See you at the event"
    ],
    'label': [
        1, 1, 1,
        0, 0, 0,
        1, 0
    ]
}

df = pd.DataFrame(data)
print(df)
```

# **Import Libraries**

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# **Feature Extraction (Bag-of-Words)**

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['text'])

y = df['label']
```

# **Data Split**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

# **Train Naive Bayes Model**

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

# **Model Testing**

```python
y_pred = model.predict(X_test)

print("Predictions:", y_pred)
print("Actual:", y_test.values)
```

# **Model Evaluation**

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

# **Model Saving**

```python
import joblib

# Save trained model
joblib.dump(model, 'naive_bayes_model.pkl')

# Save vectorizer also (VERY IMPORTANT)
joblib.dump(vectorizer, 'vectorizer.pkl')
```

# **Load Model**

```python
# Load model
loaded_model = joblib.load('naive_bayes_model.pkl')

# Load vectorizer
loaded_vectorizer = joblib.load('vectorizer.pkl')
```

# **Model Deployment**

```python
text = ["Free offer now"]

text_vec = loaded_vectorizer.transform(text)
prediction = loaded_model.predict(text_vec)

print("Prediction:", prediction[0])
```

# ***Case Study - 2 : News Group Dataset***

# **Import Libraries**

```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

# **Load Dataset**

```python
from sklearn.datasets import fetch_20newsgroups

categories = ['sci.space', 'rec.sport.hockey']
data = fetch_20newsgroups(subset='all', categories=categories)

texts = data.data
labels = data.target
for i in range(3):
    print("Text:", texts[i][:200])
    print("Label:", labels[i])
    print("------")
```

# **Text Preprocessing**

```python
import re

def preprocess(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # remove punctuation
    return text

texts = [preprocess(t) for t in texts]
```

# **Feature Extraction**

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)
```

# **Data Split**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.2, random_state=42
)
```

# **Model Training**

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

# **Model Testing**

```python
y_pred = model.predict(X_test)

print("Predictions:", y_pred)
print("Actual:", y_test)
```

# **Model Evaluation**

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

# **Model Saving**

```python
import joblib

# Save trained model
joblib.dump(model, 'naive_bayes_model_news.pkl')

# Save vectorizer also (VERY IMPORTANT)
joblib.dump(vectorizer, 'vectorizer_news.pkl')
```

# **Load Model**

```python
# Load model
loaded_model = joblib.load('naive_bayes_model_news.pkl')

# Load vectorizer
loaded_vectorizer = joblib.load('vectorizer_news.pkl')
```

# **Model Deployment**

```python
target_names = data.target_names

test_samples = [
    "NASA launched a new satellite into space",
    "The hockey team won the championship game",
    "3D rendering and computer graphics techniques improved"
]

test_vec = loaded_vectorizer.transform(test_samples)
predictions = loaded_model.predict(test_vec)

for text, pred in zip(test_samples, predictions):
    print("\nText:", text)
    print("Predicted Category:", target_names[pred])
```
