```python
data=["Fruits and vegetables are good for health.",
"Technology makes our life easier.",
"Mobile phones are very useful in daily life.",
"The teacher explained the lesson clearly.",
"I understood the topic after the class.",
"Music helps me relax after a long day.",
"Listening to songs makes me feel happy.",
"The car stopped working on the road.",
"The vehicle broke down in the middle of the street.",
"I love drinking coffee in the morning.",
"Tea is my favorite drink during winter.",
"The exam was difficult this semester.",
"Many students found the test very hard",
"I like to read books in my free time.",
"Reading books helps me improve my vocabulary.",
"The weather is very hot today.",
"It is sunny and warm outside.",
"I enjoy playing football with my friends.",
"We won the football match yesterday.",
"Eating healthy food keeps the body strong."]
```

```python
import pandas as pd

pairs = [
    (data[7], data[8]),
    (data[11], data[12]),
    (data[15], data[16]),
    (data[13], data[14]),
    (data[0], data[19]),
    (data[0], data[1]),
    (data[2], data[10]),
    (data[3], data[18]),
    (data[5], data[6]),
    (data[17], data[18])
]

df_pairs = pd.DataFrame(pairs, columns=["Sentence1", "Sentence2"])
df_pairs
```

```python
pairs = []

for i in range(0, len(data)-1, 2):
    pairs.append((data[i], data[i+1]))

df_pairs = pd.DataFrame(pairs, columns=["Sentence1", "Sentence2"])
df_pairs
```

```python
import numpy as np
import string

import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

from nltk.corpus import wordnet as wn

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
import re
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)

df_pairs["Clean1"] = df_pairs["Sentence1"].apply(preprocess)
df_pairs["Clean2"] = df_pairs["Sentence2"].apply(preprocess)
```

```python
vectorizer = TfidfVectorizer()

combined = df_pairs["Clean1"].tolist() + df_pairs["Clean2"].tolist()
vectorizer.fit(combined)

vec1 = vectorizer.transform(df_pairs["Clean1"])
vec2 = vectorizer.transform(df_pairs["Clean2"])
```

```python
cosine_scores = []

for i in range(len(df_pairs)):
    score = cosine_similarity(vec1[i], vec2[i])[0][0]
    cosine_scores.append(score)

df_pairs["Cosine_Similarity"] = cosine_scores
df_pairs.head()
```

```python
def jaccard_similarity(s1, s2):
    set1 = set(s1.split())
    set2 = set(s2.split())
    return len(set1 & set2) / len(set1 | set2)

df_pairs["Jaccard_Similarity"] = df_pairs.apply(
    lambda x: jaccard_similarity(x["Clean1"], x["Clean2"]), axis=1
)
```

```python
def wordnet_similarity(word1, word2):
    syn1 = wordnet.synsets(word1)
    syn2 = wordnet.synsets(word2)

    if syn1 and syn2:
        return syn1[0].wup_similarity(syn2[0])
    return 0
```

```python
def sentence_wordnet_similarity(s1, s2):
    words1 = s1.split()
    words2 = s2.split()

    scores = []

    for w1 in words1:
        for w2 in words2:
            sim = wordnet_similarity(w1, w2)
            if sim is not None:
                scores.append(sim)

    if len(scores) > 0:
        return sum(scores) / len(scores)
    return 0

df_pairs["WordNet_Similarity"] = df_pairs.apply(
    lambda x: sentence_wordnet_similarity(x["Clean1"], x["Clean2"]), axis=1
)

df_pairs
```
