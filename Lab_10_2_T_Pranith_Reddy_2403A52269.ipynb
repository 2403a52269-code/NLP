IMPORTING THE **LIBRARIES**

```python
from gensim.models import KeyedVectors
import gensim.downloader as api
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

**IMPORTING WORD2VEC AND GLOVE MODELS**

```python
model_word2vec = api.load("word2vec-google-news-300")
model_glove = api.load("glove-wiki-gigaword-100")
```

```python
word_list = [

    # Sports
    "football", "cricket", "tennis", "basketball", "hockey",

    # Professions
    "engineer", "lawyer", "architect", "pilot", "chef",

    # Fruits
    "apple", "banana", "mango", "orange", "grape",

    # Colors
    "red", "blue", "green", "yellow", "purple",

    # Weather
    "rain", "snow", "storm", "wind", "cloud",

    # Space
    "planet", "star", "galaxy", "asteroid", "comet",

    # Emotions
    "happy", "sad", "angry", "excited", "nervous",

    # Household Items
    "table", "chair", "sofa", "bed", "mirror",

    # Music
    "guitar", "piano", "drum", "violin", "flute"
]
```

```python
word_list = [

    # Sports
    "football", "cricket", "tennis", "basketball", "hockey",

    # Professions
    "engineer", "lawyer", "architect", "pilot", "chef",

    # Fruits
    "apple", "banana", "mango", "orange", "grape",

    # Colors
    "red", "blue", "green", "yellow", "purple",

    # Weather
    "rain", "snow", "storm", "wind", "cloud",

    # Space
    "planet", "star", "galaxy", "asteroid", "comet",

    # Emotions
    "happy", "sad", "angry", "excited", "nervous",

    # Household Items
    "table", "chair", "sofa", "bed", "mirror",

    # Music
    "guitar", "piano", "drum", "violin", "flute"
]
```

```python
vector_word2vec = [model_word2vec[word] for word in word_list if word in model_word2vec]
vector_word2vec = np.array(vector_word2vec)
vector_word2vec
print("vector shape:")
print(vector_word2vec.shape)

print(vector_word2vec[0])
print("word list:")
print(word_list)
```

```python
vector_glove = [model_glove[word] for word in word_list if word in model_glove]
vector_glove = np.array(vector_glove)
vector_glove
print("vector shape:")
print(vector_glove.shape)
print(vector_glove)
print("word list:")
print(word_list)
```

```python
tsne=TSNE(n_components=2,random_state=42,perplexity=10,learning_rate=100)

tsne_word2vec=tsne.fit_transform(vector_word2vec)

print("original shape:")
print(vector_word2vec.shape)
print("tsne shape:")
print(tsne_word2vec.shape)

print(tsne_word2vec[:5])
```

**FIGURE OF TSNE_WORD2VEC**

```python
plt.figure(figsize=(8,6))
plt.scatter(tsne_word2vec[:,0], tsne_word2vec[:,1])

for i, word in enumerate(word_list):
    plt.annotate(word, (tsne_word2vec[i,0], tsne_word2vec[i,1]))

plt.show()
```

```python
tsne=TSNE(n_components=2,random_state=42,perplexity=10,learning_rate=100)

tsne_glove=tsne.fit_transform(vector_glove)

print("original shape:")
print(vector_glove.shape)
print("tsne shape:")
print(tsne_glove.shape)
print(tsne_glove)
```

**FIGURE OF TSNE_GLOVE**

```python
plt.figure(figsize=(8,6))
plt.scatter(tsne_glove[:,0], tsne_glove[:,1])
for i, word in enumerate(word_list):
  plt.annotate(word, (tsne_glove[i,0], tsne_glove[i,1]))

plt.show()
```
